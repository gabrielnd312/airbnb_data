###**Outliers**

By the way the histogram results were presented, it is possible to presume the presence of *outliers* in the considered data.

For example, the variance in the `price` and `minimum_nights` columns

The values ​​do not follow a uniform distribution, and distort the entire graphical representation with some extremely high values, which affect our average. To confirm, there are two quick ways to help detect *outliers*.

We use:

* Statistical summary via the `describe()` method
* Plot `boxplots` graphs for the variable.

```
#Check the statistical summary of numerical variables
df[["price_usd","minimum_nights","availability_365"]].describe()
```

When looking at the statistical summary, it is immediately possible to verify some possible inconsistencies with values, indicating unrealistic and/or incorrect numbers.

 Which would lead to a **compromised analysis** if considered:


---


 Accommodation values ​​in **[price_usd]** -> The average is **189 USD**, with **75% of the data being under USD 214**.
 
 
 However, the dataset indicates values ​​of up to **U$ 9399 per night**. We know that this is not a pattern that repeats itself frequently in the dataset, and considering it would only increase the other averages, deviating the data from the demonstration of reality.



---
The values ​​indicated in **[minimum_nights]** -> *75%* of the data in the dataset indicate that the **minimum number of nights to stay is up to 2 nights**. As a maximum amount, we note that it is possible to reach the number of **1,000 nights** minimum, which would be more than 2 years of minimum stay.

For a service like Airbnb *(and even a normal rental contract)* we know that this value does not correspond to reality, and it can also increase our average upwards if considered in the analysis as a real value.


---

###**Boxplot**

For a better visualization of the Outliers and their effects on our analysis, we used the Boxplot in the order below:

Boxplot for column **[price_usd]**:



```
#Code for boxplot chart [price_usd]
df.price_usd.plot(kind="box", vert = False, figsize = (15 , 3))
plt.show()

#Check accommodations with a minimum price above $600:
print("Values ​​above 600 dollars:")
print("{} entries.".format(len(df[df.price_usd > 600])))
print("Percentage of entries over $600: {:.4f}%".format((len(df[df.price_usd > 600]) / df.shape[0]) * 100))

print("-=" * 30)

#Check accommodations with a minimum price above $800:
print("Values ​​above 800 dollars:")
print("{} entries.".format(len(df[df.price_usd > 800])))
print("Percentage of tickets over $800: {:.4f}%".format((len(df[df.price_usd > 800]) / df.shape[0]) * 100))
```

Initially, the cut-off point was set at **U$600.00** per night, however, the representation of this value in our dataset exceeds **2.7%** of entries with **277 accommodations.**

Taking its relevance into account, the cutoff for accommodations above **U$800.00** per night was changed, with only **97** entries representing less than **1%** in our dataset.

##**Histograms without Outliers**

According to the analysis carried out in the columns identified above *([price_usd] and [minimum_nights])*, the relevant presence of Outliers was demonstrated, enough to deviate the analysis from reality.

Below we clean the DataFrame to perform a new analysis of its histograms and check the previous effects of the Outliers.

```
#Creation of DataFrame.clean() removing Outliers
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price_usd > 800].index, axis = 0, inplace = True)
df_clean.drop(df_clean[df_clean.minimum_nights > 30]. index, axis =0, inplace = True)
```

Earlier we identified that the column **[neighbourhood_group]** is completely empty. Therefore, it will not be useful in our analysis.

We also remove the original column **[price]** so that it does not confuse with the new column of converted and cleaned values ​​of Outliers.

Follow the removal below:

```
df_clean.drop("neighbourhood_group", axis = 1, inplace = True)
df_clean.drop("price", axis = 1, inplace = True)
```

##**Finally, we plot the histograms after cleaning up.**

```
df_clean.hist(bins = 15, figsize = (15, 10));
```

### **Q6. What is the most expensive location in the dataset?**

To identify this information more accurately and quickly, we can use *groupby ()*.

In this way, we can carry out our price analysis based on the location of the accommodation.


```
# Identification of prices by neighborhoods, on average
neighbourhood_mean = df_clean.groupby(["neighbourhood"]).price_usd.mean().sort_values(ascending = False) [:10]
print(neighbourhood_mean)
```

```
#Plotting the price chart


# Group the DataFrame by neighborhood and calculate average prices
neighbourhood_mean = df_clean.groupby("neighbourhood")["price_usd"].mean()

neighbourhood_mean = neighbourhood_mean.sort_values(ascending=False)

# Select only the 10 most expensive neighborhoods
top_10_neighbourhoods = neighbourhood_mean.head(10)


# Define the minimum and maximum point of the Y axis
plt.ylim(100,250)

# Plot the bar chart
plt.bar(top_10_neighbourhoods.index, top_10_neighbourhoods)
plt.xlabel("Neighbourhood")
plt.ylabel("Average Price (USD)")
plt.title("Top 10 most expensive neighbourhoods")
plt.xticks(rotation = 90)
plt.show()
```

#**Interactive map using Folium.**

<left><img alt="Tk/Jp" width="30%" src="https://images.unsplash.com/photo-1585157603881-3bfc18eb2ed1?ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D&auto=format&fit=crop&w=378&q=80"></left>

```
#Creating a smaller dataset for better map visualization:
df_map = df_clean.sample(frac=0.05, random_state=0)

print(df_map.shape)
```

To plot our map, we will use the information from the **latitude, longitude** columns. To be able to use them, let's turn them into variables first.

We will also use the information in the **["name"]** column for our interactive map. This information will appear when we hover the mouse over the accommodation properties.

Once this is done, everything will be ready to create the map.



```
#Creating the variables

latitude = df_map.latitude.values
longitude = df_map.longitude.values
listings = df_map.name.values
```

**Creating the map**

<left><img alt="Tk/Jp" width="50%" src="https://images.unsplash.com/photo-1553290322-0440fe3b1ddd?ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D&auto=format&fit=crop&w=1170&q=80"></left>

```

#Map
jpn = folium.Map(location=[35.6648321,139.7347197],
                 zoom_start=10)

#Adding the bookmarks
for lat, lon, listing in zip(latitude, longitude, listings):
  folium.Marker(location=[float(lat), float(lon)], popup=listing, language = "en").add_to(jpn)



#Plotting the map
jpn.save("jpn_map.html")
jpn
```
